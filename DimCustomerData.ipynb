```python
# Import necessary libraries
import csv
import random
from faker import Faker

# Initialize Faker
fake = Faker()

# Input the number of rows the CSV file should have
num_rows = int(input("Enter the number of rows the CSV file should have: "))

# Input the name of the CSV file (e.g., data.csv)
csv_file = input("Enter the name of the CSV file (e.g., data.csv): ")

# Create the header
header = ['First Name', 'Last Name', 'Gender', 'DateOfBirth', 'Email', 'Phone Number', 
          'Address', 'City', 'State', 'Postal Code', 'Country', 'LoyaltyProgramID']

# Open the CSV file and write data
with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    # Write header to the CSV file
    writer.writerow(header)

    # Loop and generate multiple rows
    for _ in range(num_rows):
        row = [
            fake.first_name(),
            fake.last_name(),
            random.choice(['M', 'F', 'Others', 'Not Specified']),
            fake.date_of_birth(minimum_age=18, maximum_age=90).strftime("%Y-%m-%d"),  # Ensures realistic DoB
            fake.email(),
            fake.phone_number(),
            fake.address().replace(",", " ").replace("\n", " "),  # Ensures address formatting
            fake.city(),
            fake.state(),
            fake.postcode(),
            fake.country(),
            random.randint(1000, 9999)  # More realistic Loyalty Program ID
        ]

        # Write row to CSV
        writer.writerow(row)

# Display success message in Jupyter Notebook
print(f"CSV file '{csv_file}' with {num_rows} rows has been created successfully.")

# Verify the first few rows by reading the file (for Jupyter Notebook display)
import pandas as pd

df = pd.read_csv(csv_file)
df.head()  # Display the first few rows in the notebook
```
